```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Описание системы уравнений
def model(phi, t, Omega, A):
    return Omega - A * np.sin(phi)
```

```python
phi0 = 0
Omega_values = np.linspace(1, 10, 10)
A_values = np.linspace(1, 10, 10)

t = np.linspace(0, 100, 1000)
for A in A_values:
  for Omega in Omega_values:
    solution = odeint(model, phi0, t, args=(Omega, A))
    plt.plot(t, solution % (2 * np.pi))  # берем значение по модулю 2π
    plt.title(f'Решение дифференциального уравнения для нейрона A = {int(A)} Omega = {int(Omega)}')
    plt.xlabel('Время')
    plt.ylabel('Phi')
    plt.grid()
    plt.show()
```
