```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Задание 1.1


```python
def modelFerhulst(r, x0, n):
  population = [(1 + r) * x0 - r * x0**2]
  for i in range(n - 1):
    population.append((1 + r)*population[-1] - r * population[-1]**2)
  plt.plot(range(1, n + 1), population)
  plt.title('График Xn')
  plt.xlabel('n')
  plt.ylabel('Xn')
  plt.grid()
  plt.show()

  plt.plot(population[1:], population[:-1])
  plt.title('Фазовый портрет')
  plt.xlabel('Xn')
  plt.ylabel('Xn+1')
  plt.grid()
  plt.show()


  df = pd.DataFrame({'Xn+1':population[1:], 'Xn':population[:-1]}, index = range(1, n))
  #with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(df)
```

Задание 3

```python
def modelDoloto(r, G0, n):
  G = [r * G0 * (1 - G0)]
  for i in range(n):
    G.append(r * G[-1] * (1 - G[-1]))
  plt.plot(range(1, n + 1), G)
  plt.title('График Gn')
  plt.xlabel('n')
  plt.ylabel('Gn')
  plt.grid()
  plt.show()

  plt.plot(G[1:], G[:-1])
  plt.title('Фазовый портрет')
  plt.xlabel('Gn')
  plt.ylabel('Gn+1')
  plt.grid()
  plt.show()
  df = pd.DataFrame({'Gn+1':G[1:], 'Gn':G[:-1]}, index = range(1, n))
  #with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(df)
```

```python
modelDoloto(r = 0.5, G0 = 0.002, n = 70)
```

При значении r < 1 осевая нагрузка не доводится до долота.


```python
modelDoloto(r = 1.5, G0 = 0.002, n = 70)
```

При 1 < r < 3 приближается к конечному состоянию G.

```python
modelDoloto(r = 3.5, G0 = 0.002, n = 70)
```

При  3 < r < 3.5 начинает периодически осциллировать между двумя точками.

```python
modelDoloto(r = 3.7, G0 = 0.002, n = 70)
```

При r > 4 перестает быть периодическим, поведение превращается в
хаотичное.
