```python
import pandas as pd
import requests
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LassoCV, Ridge, ElasticNet, LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
url='https://drive.google.com/uc?id=1LBDnhITL0Wqwp5G6M6IBI-SSz8BIoNec'
df = pd.read_csv(url)
```

```python
df.info()
```

```python
X = df[['SkinThickness', 'BMI']]
y = df[['Insulin']]
```

```python
df.corr()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

Lasso

```python
lasso_regression = Lasso(alpha=0.1)
lasso_regression.fit(X_train, y_train)
print('R2 train:', lasso_regression.score(X_train, y_train))
print('R2 test:', lasso_regression.score(X_test, y_test))
```

```python
lasso_regression.coef_
```

Ridge

```python
ridge_regression = Ridge(alpha=0.1)
ridge_regression.fit(X_train, y_train)
print('R2 train:',ridge_regression.score(X_train, y_train))
print('R2 test:',ridge_regression.score(X_test, y_test))
```

ElasticNet

```python
elastic_regression = ElasticNet(alpha=0.1)
elastic_regression.fit(X_train, y_train)
print('R2 train:',elastic_regression.score(X_train, y_train))
print('R2 test:',elastic_regression.score(X_test, y_test))
```

Polynom 2

```python
poly_regression = PolynomialFeatures(degree=2)
X_train_poly = poly_regression.fit_transform(X_train)
X_test_poly = poly_regression.fit_transform(X_test)

linear_regression = LinearRegression()
linear_regression.fit(X_train_poly, y_train)
print('R2 train:',linear_regression.score(X_train_poly, y_train))
print('R2 test:',linear_regression.score(X_test_poly, y_test))
```
