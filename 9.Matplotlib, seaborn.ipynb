```python
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4
data=np.array([[3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
               [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
               [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
               [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]])
df = pd.DataFrame(data = np.transpose(data), index = people)
ax = df.plot.barh(stacked = True, xlabel="Scores", color=['red', 'green', 'white', 'purple'], legend=False, edgecolor='k', width=0.8)

plt.tight_layout()
for i, p in enumerate(ax.patches):
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    percent = int(data.ravel()[i]/df.iloc[:, i // 8].sum() * 100)
    ax.annotate(f'{percent}%\n', (x + width/2, y + height/2), ha='center',va='center')
plt.show()
```
