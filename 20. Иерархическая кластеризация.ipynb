```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
url='https://drive.google.com/uc?id=1LBDnhITL0Wqwp5G6M6IBI-SSz8BIoNec'
df = pd.read_csv(url)
```

№1


```python
X = df[['BMI', 'Age']]
```

```python
y_pred = KMeans(n_clusters=6, n_init='auto').fit_predict(X)
```

```python
fig, ax = plt.subplots()
sc = ax.scatter(X['BMI'], X['Age'], c = y_pred, s=5)
ax.legend(*sc.legend_elements(), title='Clusters')
plt.xlabel("BMI", fontsize=14)
plt.ylabel("Age", fontsize=14)
plt.show()
```

```python
fig = plt.figure(figsize=(16, 30))
fig.patch.set_facecolor('white')

mergings = linkage(X,  method='ward')

R = dendrogram(mergings, labels = y_pred, orientation = 'left', leaf_font_size = 12)

plt.show()
```
